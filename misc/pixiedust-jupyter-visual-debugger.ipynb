## How to use it

1) Import pixiedust module before using it.

```python
import pixiedust
```

2) Add cell magic **%%"pixie_debugger" as first line** in the cell you're going to debug. 

```python
%%pixie_debugger
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def speak(self):
        uppercase = self.name.upper()
        print(f'Hi my name is {uppercase} and I\'m {self.age}')

john = Person("John", 36)
john.speak()
```

# WARING: 

The debugger is still under active development, so be patient if sometime you have to restart the kernel and **remember to save your model** before using it or you may have to train it again ;-)

From my experience it works very well with not so complex code, but has some problem debugging very complex code (expecially with nested/recursive loops).
If you're going to debug an "error metric" or a "data loader", it's better to extract it from the context (passing expected inputs) instead of debugging it in the train loop.
